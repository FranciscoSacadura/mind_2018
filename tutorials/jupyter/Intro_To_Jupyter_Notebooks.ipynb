<h1><center>I Wanna Be Interactive</center></h1>

- Rather than keeping code *scripts* code and *execution* separate, jupyter integrates both together using the concept of **code cells**  
- Cells pair "snippets" of code with the output obtained from running them  
- Can contain plots/figure inline alongside code required to generate them
- Access to the file system  
- "IDE" for analysts (+jupyter lab+)
